```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import os
import matplotlib.pyplot as plt
pd.set_option('display.max_column',190)
pd.set_option('display.max_rows',190)
```

```python
def calc_null_percentage_job(df,threshold):
    """for each col of the data 
            count the number of NA
            if the  number of NA exceed threshold 
                append to a list
            else 
                continue"""
    col_drop=[]
    num_of_rows=len(df)
    threshold_num=(threshold/100) * num_of_rows
    for column in df:
        if df[column].isna().sum() > threshold_num:
            col_drop.append(column)
    return col_drop

def remove_from_csv(cols,df):
    df=df.drop(columns = cols,axis=1)
    return df
```

```python
test_files = os.listdir('Scale_down')
test_files
```

```python
training_df = pd.DataFrame()
for file in test_files[:-1]:
    name = f'Scale_down/{file}'
    temp = pd.read_csv(name, index_col=0)
    temp.reset_index(inplace =True)
    col_drop=calc_null_percentage_job(temp,30)
    temp_remove_NA = remove_from_csv(col_drop, temp)
    print(temp_remove_NA.shape)
    training_df = pd.concat([training_df, temp_remove_NA])
    print(f'{file} Done concatenating')
```

Dealing with 7 days

```python
def failure_fn(row):
    if row.failure==1:
        return row.date- np.timedelta64(7,'D')
    else:
        return row.date
```
